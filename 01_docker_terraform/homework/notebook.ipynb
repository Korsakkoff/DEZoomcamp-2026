```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
dtype = {
    "LocationID": "Int64",
    "Borough": "string",
    "Zone": "string",
    "service_zone": "string"
}
```

```python
df = pd.read_csv('taxi_zone_lookup.csv', dtype=dtype)
```

```python
len(df)
```

```python
engine = create_engine('postgresql://postgres:postgres@db:5432/ny_taxi')
```

```python
df.to_sql(
    name="dim_taxi_zone",
    con=engine,
    if_exists="replace",
    index=False
)

print("Loaded ", len(df), " rows.")
```
