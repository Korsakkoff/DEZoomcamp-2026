```python
import pandas as pd
import pyarrow
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@db:5432/ny_taxi')
```

```python
df = pd.read_parquet(
   'green_tripdata_2025-11.parquet'
)

df.dtypes
```

```python
df["store_and_fwd_flag"] = (
    df["store_and_fwd_flag"]
    .map({"Y": True, "N": False})
    .astype("boolean")
)

df = df.astype({
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "payment_type": "Int64",
    "trip_type": "Int64"
})
```

```python
df.dtypes
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata', con=engine))
```

```python
df.to_sql(
    name="green_tripdata",
    con=engine,
    if_exists="replace",
    index=False
)

print("Loaded ", len(df), " rows.")
```
